```python
def data_processing(filepath):
    data = pd.read_csv(filepath)
    data = pd.get_dummies(data, columns=['wd'])
    data_concentration = data.dropna()
    data_prenormalized = data_concentration.drop(['No','year','month','day','hour','PM10','station'],axis = 1)
    data_min = np.min(data_prenormalized['PM2.5'])
    data_max = np.max(data_prenormalized['PM2.5'])
    data_normalized = data_prenormalized.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
    return data_min,data_max,data_normalized
```

```python
#data preprocessing for all districts
shunyi_min,shunyi_max,shunyi_data = data_processing('data/PRSA_Data_Shunyi_20130301-20170228.csv')
changping_min,changping_max,changping_data = data_processing('data/PRSA_Data_Changping_20130301-20170228.csv')
huairou_min,huairou_max,huairou_data = data_processing('data/PRSA_Data_Huairou_20130301-20170228.csv')
dongcheng_min,dongcheng_max,dongcheng_data = data_processing('data/PRSA_Data_Tiantan_20130301-20170228.csv')
chaoyang_min,chaoyang_max,chaoyang_data = data_processing('data/PRSA_Data_Aotizhongxin_20130301-20170228.csv')
```

# 2.RandomForest Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
def valid(col,drop_cols):
    return col not in drop_cols

def gen_train_test_data(data,n,dependent_col,drop_cols):
    X = data[[x for x in data.columns if valid(x,drop_cols)]]
    y = data[dependent_col]
    if n>0:
        X = X.iloc[:-n]
        y = y.shift(-n).dropna()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    return X_train, X_test, y_train, y_test
```

```python
drop_cols = ['No','year','day','hour','PM2.5','PM10','station']
dependent_col = 'PM2.5'
X_train, X_test, y_train, y_test = gen_train_test_data(shunyi_data,1,dependent_col,drop_cols)
```

```python
def model_tunning():
    criterion=['mse','mae']
    n_estimators = [200,300,400,500,600]
    max_features = [8,11,15,20]
    max_depth = [10,20,30,40,50]
    max_depth.append(None)
    min_samples_split = [1.0,2,3,4,5]
    min_samples_leaf = [1, 2, 4]
    bootstrap = [True, False]
    random_grid = {'criterion':criterion,
                  'n_estimators': n_estimators,
                  'max_features': max_features,
                  'max_depth': max_depth,
                  'min_samples_split': min_samples_split,
                  'min_samples_leaf': min_samples_leaf,
                  'bootstrap': bootstrap}

    clf= RandomForestRegressor()
    clf_random = RandomizedSearchCV(estimator=clf, param_distributions=random_grid,
                              n_iter = 10,  
                              cv = 3, verbose=2, random_state=42, n_jobs=1)
    
    clf_random.fit(X_train, y_train)
    print (clf_random.best_params_)
```

```python
model_tunning()
```

```python
rf_model = RandomForestRegressor()
# Train the model
rf_model.fit(X_train, y_train)
print("without tuning:")
print(rf_model.score(X_test, y_test))
```

```python
# Create a model 
rf_model1 = RandomForestRegressor(oob_score=True,n_estimators=300, min_samples_split=4, min_samples_leaf=1, max_features=11, max_depth=None, criterion='mse', bootstrap=True)
# Train the model
rf_model1.fit(X_train, y_train)
print(rf_model1.score(X_test, y_test))
```

```python
print("predicted data and real data:")
print(pd.concat([pd.Series(rf_model1.predict(X_test)), y_test.reset_index(
drop=True)], axis=1))
```

```python
#get test data
def get_test_data(data,n,dependent_col,drop_cols):
    X = data[[x for x in data.columns if valid(x,drop_cols)]]
    y = data[dependent_col]
    if n>0:
        X = X.iloc[:-n]
        y = y.shift(-n).dropna()
    return X.iloc[-24:],y.iloc[-24:]
```

```python
X_test_shunyi,y_test_shunyi = get_test_data(shunyi_data,1,dependent_col,drop_cols)
X_test_shunyi,y_test_shunyi
```
